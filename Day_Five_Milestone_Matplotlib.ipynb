```python
import numpy as np 
import pandas as pd
import altair as at
import seaborn as sb 
import matplotlib.pyplot as mt
import os 
from pathlib import Path
```

```python
Pth = '//Users//ercagpince//Downloads//Pandas_Milestone'
Init_L = os.listdir(Pth)
FileList = list(range(2))
FileList[0] = "Final_DF.csv" #[x for x in Init_L if x.endswith('csv')]
FileList[1] = "Value_of_Energy_Cost_Savings_Program_Savings_for_Businesses_-_FY2020.csv"
Complete_File = Path(Pth, FileList[0])
Data = pd.read_csv(Complete_File)
Old_Data = pd.read_csv(Path(Pth,FileList[1]))
```

```python
## Average total savings vs. Total jobs created plot in linear scale 
#fig, axes = mt.subplots(1, 2, figsize=(15, 5), sharey=True)
sb.set_theme()
plt_1 = sb.relplot(data = Data, \
                 x = "Average total savings", \
                     y ="Total jobs created")
plt_1.set(xscale = 'linear')
mt.savefig(Pth + "/" + "Average_total_saving_linear.pdf")
```

```python
plt_2 = sb.relplot(data = Data, \
                 x = "Average total savings", \
                     y ="Total jobs created") 
plt_2.set(xscale = 'log')
mt.savefig(Pth + "/" + "Average_total_saving_log.pdf")
```

```python
plt_3 = sb.displot(data = Data,\
                   x = "Average total savings",\
                       log_scale = True)

plt_3.set_axis_labels("Average total savings ($)")

mt.savefig(Pth + "/" + "Hist_Average_total_savings_log.pdf")
```
