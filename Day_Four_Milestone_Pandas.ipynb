```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as mt
import os
import seaborn as sb 
from pathlib import Path
```

```python
Pth = '//Users//ercagpince//Downloads//Pandas_Milestone'

Init_L = os.listdir(Pth)
FileList = [x for x in Init_L if x.endswith('csv')]
Complete_File = Path(Pth, FileList[1])
Data = pd.read_csv(Complete_File)
```

```python
C_names = pd.unique(Data['Company Name'])
TotalNumber = C_names.shape[0]
print('Total number of companies:',TotalNumber)
```

```python
Queens_Jobs = Data.loc[Data["Borough"] == "Queens", "Job created"]
Number_QJ = Queens_Jobs.sum(skipna = True) 
print('Total number of jobs created in Queens:', Number_QJ)
```

```python
Emails = Data["company email"]
Sep_Emails = Emails.str.split(pat = '@') 

New = Sep_Emails[~Sep_Emails.isnull()]
New.reset_index(drop=True,inplace=True)
Domains = New.map(lambda x: x[-1])
##
Total_uni_domain = np.size(Domains.unique())
##
print('The total number of the unique domain names that companies use:', Total_uni_domain)
```

```python
Final_DF = pd.DataFrame(columns = ["NTA","Average total savings","Total jobs created"])
Temp = pd.DataFrame(columns = ["NTA","Average total savings","Total jobs created"])

Subset = Data[["Industry","NTA"]]
NTAs = Subset.NTA.dropna().unique()

for i in range(NTAs.shape[0]): 
    Subset = Data.loc[Data["NTA"] == NTAs[i]]
    n_entry = Subset.shape[0] 
    if n_entry > 4:
       Mean_Save =  Subset["Total Savings"].mean() 
       Total_Jobs = Subset["Job created"].sum() 
       NTA = Subset["NTA"].unique()
       
       Temp["NTA"] = NTA
       Temp["Average total savings"] = Mean_Save
       Temp["Total jobs created"] = Total_Jobs
       
       Final_DF = pd.concat([Final_DF, Temp], axis=0)

#Show the final dataframe before saving it 
Final_DF
```

```python
#Save the dataframe to .CSV file in the target folder 
    
Export_File = Path(Pth, 'Final_DF.csv')
Final_DF.to_csv(Export_File)
```
